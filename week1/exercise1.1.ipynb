<h3> (f) What daily expenditures would you predict for a new client of fifty years old? And for someone who is twenty-five years old?

Best estimate without knowledge of simple regression would be the mean. <br>
For age 50 (>= 40), prediction is 95.8. <br>
For age 25 (< 40), prediction is 106.4.

```r
above40 <- data[data$Age >= 40, ]
fit <- lm(Expenditures ~ Age, data=above40)
pred <- predict(fit, newdata = data.frame(Age = c(50)))
sprintf("Predicted expenditure for a client of 50 years old: %f", pred)
```

```r
below40 <- data[data$Age < 40, ]
fit <- lm(Expenditures ~ Age, data=below40)
pred <- predict(fit, newdata = data.frame(Age = c(25)))
sprintf("Predicted expenditure for a client of 25 years old: %f", pred)
```
